```python
import pandas as pd

df = pd.read_csv("data/raw/sales_data.csv")
df.info()
df.head()
df["Date"].agg(["min", "max"])
df[["Store ID", "Product ID", "Category", "Region"]].nunique()
```

```python
from pathlib import Path

df.columns = df.columns.str.lower().str.replace(" ", "_").str.replace("/", "_")
df["date"] = pd.to_datetime(df["date"])
df = df[(df["inventory_level"] >= 0) & (df["price"] > 0) & (df["units_sold"] >= 0)].copy()
df["revenue"] = df["units_sold"] * df["price"] * (1 - df["discount"] / 100)

Path("data/processed").mkdir(parents=True, exist_ok=True)
df.to_csv("data/processed/retail_inventory_clean.csv", index=False)
df.shape
```
