```python
import pandas as pd

df = pd.read_csv("data/raw/marketing_mix.csv")
df.info()
df.head()
```

```python
from pathlib import Path

df["Date"] = pd.to_datetime(df["Date"])
channels = ["TikTok", "Facebook", "Google Ads"]
df[channels] = df[channels].fillna(0)  # a handful of zero-spend days logged as blank, not actually missing
df["total_spend"] = df[channels].sum(axis=1)
df["year"] = df["Date"].dt.year

Path("data/processed").mkdir(parents=True, exist_ok=True)
df.to_csv("data/processed/mmm_clean.csv", index=False)
df.shape
```
